# Merge

```python
data4={'emp_id':[101,102,103,104],
       'salary':[234,345,456,567],
       'pf':[23,24,35,67]
      }
```

```python
data5={'emp_id':[101,102,104,105],
       'mob_n0':[23434,35645,45786,59967],
       'house_nu':[243,254,635,677]
      }
```

```python
#converting dictionary to dataframe
df4=pd.DataFrame(data4)
df4
```

```python
#converting dictionary to dataframe
df5=pd.DataFrame(data5)
df5
```

```python
#common data between two dataframe
#default is inner merge
pd.merge(df4,df5)
```

```python
df4
```

```python
df5
```

```python
#left merge
pd.merge(df4,df5, how='left')
```

```python
#right merge
pd.merge(df4,df5, how='right')
```

```python
#outer merge
pd.merge(df4,df5, how='outer')
```

```python
#left merge by mentioning the column name based on which we want to merge
pd.merge(df4,df5, how='left', on='emp_id')
```

```python
data6={'emp_id1':[101,102,103,104],
       'salary':[234,345,456,567],
       'pf':[23,24,35,67]
      }
```

```python
data7={'emp_id2':[101,102,104,105],
       'mob_n0':[23434,35645,45786,59967],
       'house_nu':[243,254,635,677]
      }
```

```python
#converting dictionary to dataframe
df6=pd.DataFrame(data6)
df6
```

```python
#converting dictionary to dataframe
df7=pd.DataFrame(data7)
df7
```

```python
#even there is no common column 
#default inner join
pd.merge(df6,df7, left_on='emp_id1',right_on='emp_id2')
```

```python
#dleft join
pd.merge(df6,df7, left_on='emp_id1',right_on='emp_id2',how='left')
```

# Calculations Using UDF(user defined function) 

```python
#reading csv file
df_sales=pd.read_csv(r'D:\1.SOMALI MISHRA\1.iNeuron\Python\Data set\sales data\sales_data_final.csv')
```

```python
df_sales.head()
```

```python
#the function has been created to apply some conditions in table
def profit_flag(a):
    if a>=0:
        return 'positive'
    else:
        return 'negative'
```

```python
profit_flag(44)
```

```python
profit_flag(-44)
```

```python
#use udf(usedr defined function)
#created a new column by applying condition through udf
df_sales['flag_profit']=df_sales['profit'].apply(profit_flag)
```

```python
df_sales.head()
```

```python
#use inbuilt function
df_sales['len_cust_name']=df_sales['customer_name'].apply(len)
```

```python
df_sales.head()
```

```python
#created function to apply conditions 
def quantity_flag(a):
    if a < 10 :
        return 'low'
    elif a>10 and a<20:
        return 'medium'
    else:
        return 'high'
```

```python
quantity_flag(5)
```

```python
quantity_flag(25)
```

```python
#cretaed a new column and maintained flag on that column
df_sales['flag_quantity'] = df_sales['quantity'].apply(quantity_flag)
```

```python
df_sales.head(20)
```

```python
#using lambada function
df_sales['square_quantity'] = df_sales['quantity'].apply(lambda a : a**2)
```
