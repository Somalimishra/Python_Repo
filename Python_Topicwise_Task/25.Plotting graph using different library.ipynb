```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading a sns data,it's available inside seaborn library
df=sns.load_dataset('iris')
```

```python
df
```

```python
#default it will plot line graph
#it will plot only numerical columns
# without using df.describe() ,still we can say something about data
# we can say the variance also ,like which has more variance
#petal length variation is more
df.plot()
```

```python
#from describe also by seeing min and max value we can say variation is more in petal_length
df.describe()
```

```python
#kind means which type of graph we want
#here alpha=dfine colour
df.plot(kind='area' , alpha=0.9)
```

```python
# stack means I want to plot the data with normal scale
df.plot(kind='area' , alpha=0.9 , stacked=False)
```

```python
# there is no relation betn two column
df.plot.scatter(x='sepal_length',y='sepal_width')
```

```python
#here we can made the decision that if sepa_length is increasing then petal_length is also increasing
df.plot.scatter(x='sepal_length',y='petal_length')
```

```python
#c means color we are using here for sepal_width
#darker is the colour more is the sepal_width
df.plot.scatter(x='sepal_length',y='petal_length',c='sepal_width')
```

```python
#s is used to fix the size of the bubble 
df.plot.scatter(x='sepal_length',y='petal_length',c='sepal_width',s=100)
```

```python
#in size instead of direct no. we can give col.name also
df.plot.scatter(x='sepal_length',y='petal_length',c='sepal_width',s=df['sepal_length']*20)
```

```python
#for hexbin we have to give gridsize instead of s
df.plot.hexbin(x='sepal_length',y='petal_length',gridsize=10)
```

```python
# cmap for colourmap is used to create a fancy graph
df.plot.hexbin(x='sepal_length',y='petal_length',gridsize=10,cmap='viridis')
```

```python
# for plotting 3D graph
# without this we can't plot any 3D graph
from mpl_toolkits import mplot3d
```

```python
# it will generate 30 data in betwn range of -1 to 6 and that would be my x axis
x=np.linspace(-1,6,30)
x
```

```python
# it will generate 30 data in betwn range of -1 to 6 and that would be my y axis
y=np.linspace(-1,6,30)
y
```

```python
#we need x axis to prepare 3d graph
z=x+y
```

```python
z
```

```python
#white colour box iscreated by (projection = "3d")
#for setting the label of x,y and z
ax=plt.axes(projection='3d')
ax.plot3D(x,y,z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
#here we are trying to create 3d graph for iris data set
ax=plt.axes(projection='3d')
ax.plot3D(df['sepal_length'],df['sepal_width'],df['petal_length'])
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
ax.set_zlabel('petal_length')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(6, 15))

# Load the example car crash dataset
crashes = sns.load_dataset("car_crashes").sort_values("total", ascending=False)

# Plot the total crashes
sns.set_color_codes("pastel")
sns.barplot(x="total", y="abbrev", data=crashes,
            label="Total", color="b")

# Plot the crashes where alcohol was involved
sns.set_color_codes("muted")
sns.barplot(x="alcohol", y="abbrev", data=crashes,
            label="Alcohol-involved", color="b")

# Add a legend and informative axis label
ax.legend(ncol=2, loc="upper left", frameon=True)
ax.set(xlim=(0, 24), ylabel="country",xlabel="Automobile collisions per billion miles")
sns.despine(left=True, bottom=True)
```
