# 7. Write a sql query to find out how many unique dress that we have based on dress id

#    9. Try to find out total dress sell for individual dress id 

```python
#Find the total sales for each and every dress id
#Creating Column Total Sales for each Dress ID as there is no slaes column in data
dress_data['Sales'] = dress_data[['29/8/2013', '31/8/2013', '2/9/2013','4/9/2013','6/9/2013','8/9/2013','10/9/2013',
                                        '12/9/2013','14/9/2013','16/9/2013','18/9/2013','20/9/2013','22/9/2013','24/9/2013'
                                        ,'26/9/2013','28/9/2013','30/9/2013','2/10/2013','4/10/2013','6/10/2013',
                                        '8/10/2013','10/10/2013','12/10/2013']].agg('sum', axis=1)
```

```python
dress_data.columns
```

```python
#This will provide the rolled up sum of sales of each unique dress_id. Similarily we can do in SQL also
print(dress_data.groupby('dress_id')['Sales'].sum()) 
```

#    10. Try to find out a third highest most selling dress id 

```python
#sales column has been created in previous query
sales_data = dress_data.groupby('dress_id')['Sales'].sum().reset_index()
print(sales_data)
print(sales_data.Sales.nlargest(3).iloc[-1]) 
#Getting 75979 as 3rd largest sum
```

```python
#If we have to do in SQL then query will be
SELECT DISTINCT SALES AS 3RDHighestSales
FROM DRESS_SALES_TABLE
ORDER BY SALES DESC
LIMIT 1 
OFFSET 2;
```
